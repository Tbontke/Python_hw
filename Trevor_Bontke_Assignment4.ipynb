```python
# A function to convert  into ordianl numbers
def to_ordinal(num: int ) -> str:
    if isinstance(num, int):
        num = str(num)
        if len(num) >= 2 and num[-2] == "1":
                ordinal = num+"th"
        else:
            if num[-1] == "1":
                ordinal = num+"st"
            elif num[-1] == "2":
                ordinal = num+"nd"
            elif num[-1] == "3":
                ordinal = num+"rd"
            else:
                ordinal = num+"th"
        return ordinal
    else:
        raise ValueError("This function only works for integer value.")         
        
```

```python
# Function to split the check using a for loop
def split_the_check(number_of_friends: int = 5):
    names = []
    bills = []
    names.append(input(print(f"What is your name?")))
    bills.append(float(input(print(f"How much was your bill, {names[0]}?"))))
    print(f"${bills[0]:.2f}\n")
    for i in range(1,number_of_friends):
        names.append(input(print(f"What is the name of your {to_ordinal(i)} friend?")))
        bills.append(float(input(print(f"How much was {names[i]}'s bill?"))))
        print(f"${bills[i]:.2f}\n")
    tip_percent = input(print(f"What percent would you like to tip?"))
    print(f"%{tip_percent}\n")
    sub_total = sum(bills)
    per_person = round(sub_total/number_of_friends*(1+(float(tip_percent)/100)), ndigits=2)
    total = per_person*number_of_friends
    print(f"The total before tipping is ${sub_total:.2f}\n")
    print(f"The price per person is ${per_person:.2f}\n")
    print(f"The total after tipping is ${total:.2f}")

        


split_the_check()
```
