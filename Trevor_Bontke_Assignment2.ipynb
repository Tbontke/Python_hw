```python
#Replace pass with body of statement to complete the homework
def split_the_check(number_of_friends):

    """Trevor Bontke, tbontke@central.uh.edu

     Program #1: Sharing the bill

     212 Python, Spring 2023

     This program prompts the users to enter their name and the name of the four friends they recently had dinner with. At each prompt the user will also be asked to enter their bill amount. Finally the user is asked to define a tip percentage and the total bill is then calculated and split evenly if possible. """

    if number_of_friends != 5:
        print(f"No one cares you have {number_of_friends} because we can't use loops.\n")

    # Collecting the names and bills from each person
    name1 = input(print("What is your name?"))
    print(name1,"\n")
    bill1 = input(print("How much did you pay?"))
    # If you cast the input as a float() in line the statment won't
    # print as until after the input is made
    bill1 = float(bill1)
    print(f"${bill1:.2f}\n\n")

    name2 = input(print("What is your 1st friend's name?"))
    print(name2,"\n")
    bill2 = input(print(f"How much did {name2} pay?"))
    bill2 = float(bill2)
    print(f"${bill2:.2f}\n\n")

    name3 = input(print("What is your 2nd friend's name?"))
    print(name3,"\n")
    bill3 = input(print(f"How much did {name3} pay?"))
    bill3 = float(bill3)
    print(f"${bill3:.2f}\n\n")

    name4 = input(print("What is your 3rd friend's name?"))
    print(name4,"\n")
    bill4 = input(print(f"How much did {name4} pay?"))
    bill4 = float(bill4)
    print(f"${bill4:.2f}\n\n")

    name5 = input(print("What is your 4th friend's name?"))
    print(name5,"\n")
    bill5 = input(print(f"How much did {name5} pay?"))
    bill5 = float(bill5)
    print(f"${bill5:.2f}\n\n")


    # Asks the user what tip percentage they will pay
    tip_perc = float(input(print(f"What percentage of the total bill would you like to tip?")))
    print(f"{tip_perc:.2f}%\n")


    # Calculates the total cost of the food before the tip is added
    subtotal = bill1 + bill2 + bill3 + bill4 + bill5
    print(f"The cost before tipping is: ${subtotal:.2f}")

    # Calculates the total cost of the food plus tip
    per_person = round(subtotal/5*(1+(tip_perc/100)),2)
    total = per_person*5
    print(f"The total cost after tipping is: ${total:.2f}")
    print(f"\nThe total cost per person is: ${per_person:.2f}") 


split_the_check(5)
```
