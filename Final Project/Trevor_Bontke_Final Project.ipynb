    Its slope
    Its midpoint

    Its length ( Euclidean distance)

```python
# Defining a function to return the slope, midpoint, and length from two points
def info(a: float,b: float) -> str:
    import math

    f = lambda x : x**2 + 5*x - 7
    m = (f(b)-f(a))/(b-a)
    mid = ((b+a)/2,(f(b)+f(a)/2))
    len = math.sqrt((b-a)**2+(f(b)-f(a))**2)
    return print(f"Slope: {m}\nMidpoint: {mid}\nLength: {len}")
```

3)Integrate a function by one trapezoid

An approximation to the integral of a function f(x) over an interval [a, b] can be found by first approximating f(x) by the straight line that goes through the endpoints (a, f(a)) and (b, f(b)), and then finding the area under the straight line (which is the area of a trapezoid). The resulting formula becomes

 https://en.wikipedia.org/wiki/Trapezoidal_rule

(3a) Write a function trapezint1(f, a, b) that returns this approximation to the integral. The argument f is a Python implementation f(x) of the mathematical function f(x).

```python
# Defining a function to compute the integral using the trapezoid estimation 
def trapezint(f,a: float,b: float,n: int=1) -> float:
    d = b-a
    d_n = d/n
    t_int = 0
    # Integrating with for loop
    for i in range(0,n):
        t_int += (d_n)*(f(a+(d_n*i))+f(a+(d_n*(i+1))))/2
    return t_int
```

```python
f = lambda x : x**2 + 5*x - 7
# Calculating the integral using various numbers of trapezoid segments
n1 = trapezint(f,1,10)
n3 = trapezint(f,1,10,3)
n10 = trapezint(f,1,10,10)
n20 = trapezint(f,1,10,20)
n1000 = trapezint(f,1,10,1000)
n50000 = trapezint(f,1,10, 50000)

print(f"By increasing the number of integration sections the accuracy of the trapezoid integration can be increased\n")
print(f"When n = 1: {n1}\nWhen n = 3: {n3}\nWhen n = 10: {n10}\nWhen n = 20: {n20}\nWhen n = 1,000: {n1000}\nWhen n = 50,000: {n50000}\n")
```

4a). Detecting Topic Trends on Twitter

One of Twitter's services is the ability to track the most popular topics. For this part of the assignment, you will do something similar. Your task is to keep track of the issues identified by users with the hashtag symbol ‘#’. You will also need to count the frequency of the hashtags you found and provide a ranking of hashtags based on their frequency. The output of your script should be one file, named top_hashtags.txt, with the N most popular hashtags, where N is a parameter to your function. For example, assume this is the content of your twitter_data.txt file:

      #lebron is the best athlete of our generation

      ML 5 Demos! Lots of great stuff to come! Yes, I&#39;m excited. :) http://htmlfive.appspot.com #io2009 #googleio

      At GWT fireside chat #googleio

      @khalid0456 No, Lebron is the best #lebron

 If N is set to 2, then your script should generate a file top_hashtags.txt with the following content (note that in the case of ties the order doesn’t matter):

      #googleio 2

      #lebron 2

```python
# Defining a special class with methods to return the number of unique hastags and their frequency
class parse:
  def __init__(self, words: dict):
    self.words = words
    
  # Method to show the number of unique hashtags
  def unique(self):
    return f"There are {len(self.words)} uniquie hashtags\n\n"
  # Method to list top N most frequent hashtags
  def topN(self, n=None):
    ls = sorted(self.words.items(), key=lambda x:x[1],reverse=True)
    out = ""
    if n is None:
      for pair in ls:
        out += f'Hashtag: "{pair[0]}"\nFrequency: {pair[1]}\n\n'
        
    else:
      for i in range(n):
        out += f'Hashtag: "{ls[i][0]}"\nFrequency: {ls[i][1]}\n\n'
    
    return out
```

```python
def top_hashtag(n: int =5, filename: str="twitter_data") -> str:
    import re
    # Reading in the contents of the "twitter_data.txt" file
    with open (filename+".txt","r",encoding='utf-8') as f:
        text = f.read()

    word_dict = {}  
    # Using regex to find all hashtag words and put them in a dict
    for word in re.findall("#\w+",text.lower()):
        try:
            word_dict[word] += 1
        except:
            word_dict.update({word:1})
    # Creating an instance of the custom class
    word_freq = parse(word_dict)
    # Writing the contents to a file called "top_hashtags.txt"
    with open ("top_hashtags.txt","w",encoding='utf-8') as f:
        f.write(word_freq.unique())
        f.write(word_freq.topN(n))
```
