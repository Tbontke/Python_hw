### Section 1

Load the Stockholm temperature dataset via appropriate Numpy functions

```python
import numpy as np
# Importing the Stockholm data into a Numpy array
data = np.loadtxt("Trevor_Bontke_Assignment9_stockholm_td_adj.dat")
```

```python
# This is just for my sanity
Year = 0
Month = 1
Day = 2
Temp_Morning = 3
Temp_Afternoon = 4
Temp_Evening = 5
```

### Section 2

Generate a Numpy array of the morning temperatures in Fahrenheit. Note temperature from imported data was initially in Celsius.

```python
# Creating a new array to store the morning temperatures in farenheit
Temp_Morning_F = data[:,Temp_Morning]*9/5+32

print(f"The first recorded morning temperautre in Stockholm was {data[0,Temp_Morning]}C or {Temp_Morning_F[0]}F")
```

### Section 3

Generate a series plot of the morning temperature in Fahrenheit vs time. Label the title, legend and axes accordingly.

```python
import matplotlib.pyplot as pl

fig, Tvt = pl.subplots(figsize=(14,4))
Tvt.plot(data[:,Year]+data[:,Month]/12.0+data[:,Day]/365.25, Temp_Morning_F)
Tvt.axis('tight')
Tvt.set_title('Morning Temperatures in Stockholm')
Tvt.set_xlabel('Year')
Tvt.set_ylabel('Temperature (F)')
pl.show()
```

### Section 4

Generate a histogram of the morning temperature in Fahrenheit to reflect the temperature distribution. Label the title, legend and axes accordingly.

```python
# Defining the number of histogram bins as the square root of the number of data points
num_bins = round(np.sqrt(Temp_Morning_F.shape[0]))
print(f"The number of data bins is sqrt({Temp_Morning_F.shape[0]}) = {num_bins}")
```

```python
# Plotting a histogram of morning temperatures using sqrt(data count) bins
fig, hist = pl.subplots(figsize=(8,6))
hist.hist(Temp_Morning_F, bins=num_bins)
hist.set_title('Morning Temperatures in Stockholm')
hist.set_xlabel('Temperature (F)')
pl.show()
```
